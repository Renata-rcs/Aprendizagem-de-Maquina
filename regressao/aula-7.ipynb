### **ExercÃ­cio**

### **Ajuste e PrevisÃ£o com RegressÃ£o Linear**

Agora que vocÃª jÃ¡ viu como a **regressÃ£o linear** funciona, sua tarefa Ã© criar um modelo de **regressÃ£o linear mÃºltipla** usando **todas as features** do dataset `sales_df`, que jÃ¡ foi **prÃ©-carregado** para vocÃª.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
sales_df = pd.read_csv("advertising_and_sales_clean.csv")
sales_df
```

```python
# Tratar valores nulos (se existirem)
sales_df = sales_df.dropna()
```

```python
# Transformar colunas categÃ³ricas em numÃ©ricas
sales_df = pd.get_dummies(sales_df, drop_first=True)
```

```python
# Separar X e y
X = sales_df.drop("sales", axis=1).values
y = sales_df["sales"].values
```

```python
# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Treinar o modelo
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python

# Fazer previsÃµes
y_pred = reg.predict(X_test)

```

```python
# Avaliar
print("Predictions: {}, Actual Values: {}".format(y_pred[:2], y_test[:2]))
```

### **Desempenho da RegressÃ£o**

Agora que vocÃª ajustou um modelo (`reg`) usando todas as features do `sales_df` e fez previsÃµes dos valores de vendas, Ã© possÃ­vel avaliar o desempenho utilizando algumas **mÃ©tricas comuns de regressÃ£o**.

As variÃ¡veis `X_train`, `X_test`, `y_train`, `y_test` e `y_pred`, alÃ©m do modelo ajustado `reg`, foram todas **prÃ©-carregadas** do exercÃ­cio anterior.

Seu objetivo Ã© descobrir:

- **O quanto as features explicam a variÃ¢ncia** dos valores-alvo (das vendas) (`RÂ²`); e
- **Avaliar a capacidade do modelo de fazer previsÃµes em dados nÃ£o vistos** (`RMSE`).

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Fazer previsÃµes com os dados de teste
y_pred = reg.predict(X_test)
```

```python

# Calcular RÂ²
r_squared = reg.score(X_test, y_test)
```

```python

# Calcular RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
# Exibir os resultados
print("RÂ²: {:.4f}".format(r_squared))
print("RMSE: {:.4f}".format(rmse))
```
