# Aula 6 - Regressão - Aprendizado Supervisionado com Scykit-Learn

```python
import pandas as pd
diabetes_df = pd.read_csv("diabetes_clean.csv")
print(diabetes_df.head())
```

```python
X = diabetes_df.drop("glucose", axis=1).values
y = diabetes_df["glucose"].values
print(type(X), type(y))
```

```python
X_bmi = X[:, 4]
print(y.shape, X_bmi.shape)
```

```python
# -1 é equivalente a última linha, e 1 afirma que temos apenas uma única coluna.  
X_bmi = X_bmi.reshape(-1, 1)  # cria uma matriz 2D de uma única coluna.
print(y.shape, X_bmi.shape)
```

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_bmi, y)
plt.ylabel("Blood Glucose (mg/dl)")
plt.xlabel("Body Mass Index")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_bmi, y)
predictions = reg.predict(X_bmi)

plt.scatter(X_bmi, y)
plt.plot(X_bmi, predictions)
plt.ylabel("Blood Glucose (mg/dl)")
plt.xlabel("Body Mass Index")
plt.show()
```

# Exercício 1


```python
import pandas as pd
import numpy as np

# Carregar os dados do arquivo CSV
sales_df = pd.read_csv("advertising_and_sales_clean.csv")

# Criar X a partir da coluna 'radio'
X = sales_df['radio'].values

# Criar y a partir da coluna 'sales'
y = sales_df['sales'].values

# Redimensionar X para formato bidimensional
X = X.reshape(-1, 1)

# Imprimir o shape de X e y
print(X.shape, y.shape)
```

# Exercício 2


```python
# Import LinearRegression
from sklearn.linear_model import LinearRegression

# Create the model
reg = LinearRegression()

# Fit the model to the data
reg.fit(X, y)

# Make predictions
predictions = reg.predict(X)

# Exibir as primeiras 5 previsões
print(predictions[:5])
```

# Exercício 3

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Create scatter plot
plt.scatter(X, y, color="blue")

# Create line plot
plt.plot(X, predictions, color="red")
plt.xlabel("Radio Expenditure ($)")
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```
